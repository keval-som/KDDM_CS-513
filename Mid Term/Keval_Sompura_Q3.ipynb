```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
folder='C:/Users/Asus/OneDrive/Stevens/Sem-2/KDDM/Mid Term/'
file='hepatitis_A.csv'
data = pd.read_csv(folder+file)
data.head()
```

```python
data.describe()
```

```python
# Scatter plot for BILIRUBIN vs SGOT
plt.scatter(data['BILIRUBIN'], data['SGOT'], alpha=0.7, color='blue')
plt.title('Scatter Plot: BILIRUBIN vs SGOT')
plt.xlabel('BILIRUBIN')
plt.ylabel('SGOT')
plt.show()

# Scatter plot for BILIRUBIN vs ALBUMIN
plt.scatter(data['BILIRUBIN'], data['ALBUMIN'], alpha=0.7, color='green')
plt.title('Scatter Plot: BILIRUBIN vs ALBUMIN')
plt.xlabel('BILIRUBIN')
plt.ylabel('ALBUMIN')
plt.show()

# Scatter plot for SGOT vs ALBUMIN
plt.scatter(data['SGOT'], data['ALBUMIN'], alpha=0.7, color='red')
plt.title('Scatter Plot: SGOT vs ALBUMIN')
plt.xlabel('SGOT')
plt.ylabel('ALBUMIN')
plt.show()
```

```python
# Box plots for BILIRUBIN, SGOT, and ALBUMIN
plt.figure(figsize=(12, 6))
data[['BILIRUBIN', 'SGOT', 'ALBUMIN']].boxplot()
plt.title('Box Plots for BILIRUBIN, SGOT, and ALBUMIN')
plt.ylabel('Values')
plt.show()
```
