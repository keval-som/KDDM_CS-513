```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
headers = ["SampleCode", "ClumpThickness", "UniformityCellSize", "UniformityCellShape", "MarginalAdhesion", "SingleEpithelialCellSize", "BareNuclei", "BlandChromatin", "NormalNucleoli", "Mitoses", "Class"]
```

```python
folder='C:/Users/Asus/OneDrive/Stevens/Sem-2/KDDM/HW3/'
file='breast-cancer-wisconsin.csv'
data = pd.read_csv(folder+file, header=None,skiprows=1, names=headers,na_values='?')
data.head()
```

```python
data = data.dropna()
data
```

```python
data['Class'] = data['Class'].map({2: 'Benign', 4: 'Malignant'}).fillna(data['Class'])
data
```

```python
X = data.drop(columns=['SampleCode', 'Class'])
y = data['Class']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X = pd.DataFrame(MinMaxScaler().fit_transform(X), columns=X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_10 = KNeighborsClassifier(n_neighbors=10)
```

```python
knn_3.fit(X_train, y_train)
```

```python
knn_5.fit(X_train,y_train)
```

```python
knn_10.fit(X_train,y_train)
```

```python
y_pred_3 = knn_3.predict(X_test)
y_pred_5 = knn_5.predict(X_test)
y_pred_10 = knn_10.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
conf_matrix_3 = confusion_matrix(y_test, y_pred_3)
print(conf_matrix_3)
```

```python
conf_matrix_5 = confusion_matrix(y_test, y_pred_5)
print(conf_matrix_5)
```

```python
conf_matrix_10 = confusion_matrix(y_test, y_pred_10)
print(conf_matrix_10)
```

```python
accuracy_3 = accuracy_score(y_test, y_pred_3) * 100
print(f'Accuracy of KNN-3: {accuracy_3:.2f}%')
```

```python
accuracy_5 = accuracy_score(y_test, y_pred_5) * 100
print(f'Accuracy of KNN-5: {accuracy_5:.2f}%')
```

```python
accuracy_10 = accuracy_score(y_test, y_pred_10) * 100
print(f'Accuracy of KNN-10: {accuracy_10:.2f}%')
```

```python
from sklearn.metrics import classification_report
```

```python
classify_report_3 = classification_report(y_test, y_pred_3)
print(classify_report_3)
```

```python
classify_report_5 = classification_report(y_test, y_pred_5)
print(classify_report_5)
```
