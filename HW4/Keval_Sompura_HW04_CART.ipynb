```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
headers = ["SampleCode", "ClumpThickness", "UniformityCellSize", "UniformityCellShape", "MarginalAdhesion", "SingleEpithelialCellSize", "BareNuclei", "BlandChromatin", "NormalNucleoli", "Mitoses", "Class"]
```

```python
folder='C:/Users/Asus/OneDrive/Stevens/Sem-2/KDDM/HW4/'
file='breast-cancer-wisconsin.csv'
data = pd.read_csv(folder+file, header=None,skiprows=1, names=headers,na_values='?')
data.head()
```

```python
data['Class'] = data['Class'].map({2: 'Benign', 4: 'Malignant'}).fillna(data['Class'])
data
```

```python
X = data.drop(columns=['SampleCode', 'Class'])
y = data['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
cart_model = DecisionTreeClassifier()
cart_model.fit(X_train, y_train)
y_pred = cart_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
report = classification_report(y_test, y_pred)
print('Classification Report:')
print(report)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(50, 30), dpi=250)
plot_tree(cart_model, feature_names=X.columns, class_names=cart_model.classes_, filled=True)
plt.show() 
```
