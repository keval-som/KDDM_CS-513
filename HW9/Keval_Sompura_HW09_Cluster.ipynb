```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
folder='C:/Users/Asus/OneDrive/Stevens/Sem-2/KDDM/HW8/'
file='wisc_bc_ContinuousVar.csv'
data = pd.read_csv(folder+file)
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])
```

```python
data.head()
```

```python
X = data.drop(['diagnosis'], axis=1)
Y = data['diagnosis']
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
agglomerative = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='single')
clusters = agglomerative.fit_predict(X)
```

```python
df_cluster=pd.DataFrame({'Actual':Y,'Cluster':clusters})
cross_tab = pd.crosstab(df_cluster['Actual'], df_cluster['Cluster'])
print("Hierarchical Clustering - Cross Tabulation:")
print(cross_tab)
```

```python
df_cluster.head()
```

```python
#KMeans
from sklearn.cluster import KMeans
```

```python
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=12)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
print("Labels:")
print(labels)
```

```python
print("Centers:")
print(centers)
```

```python
df_cluster=pd.DataFrame({'Actual':Y,'Cluster':labels})
# Create a cross-tabulation
cross_tab = pd.crosstab(df_cluster['Actual'], df_cluster['Cluster'])

print(cross_tab)
```
