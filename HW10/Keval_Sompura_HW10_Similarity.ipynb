```python
import pandas as pd
from sklearn.metrics import jaccard_score
from sklearn.metrics.pairwise import cosine_similarity
```

```python
folder='C:/Users/Asus/OneDrive/Stevens/Sem-2/KDDM/HW10/'
file='Recipe.csv'
data = pd.read_csv(folder+file)
data.head()
```

```python
# Calculate Jaccard similarity for each pair of recipes
jaccard_similarities = pd.DataFrame(index=data['Recipe'], columns=data['Recipe'])

for i, recipe1 in enumerate(data['Recipe']):
    for j, recipe2 in enumerate(data['Recipe']):
        if i <= j:
            similarity = jaccard_score(data.iloc[i, 1:].astype(int), data.iloc[j, 1:].astype(int), average='binary')
            jaccard_similarities.loc[recipe1, recipe2] = similarity
            jaccard_similarities.loc[recipe2, recipe1] = similarity

jaccard_similarities
```

```python
# Calculate cosine similarity for each pair of recipes
cosine_similarities = pd.DataFrame(
    cosine_similarity(data.iloc[:, 1:]), 
    index=data['Recipe'], 
    columns=data['Recipe']
)

cosine_similarities
```
